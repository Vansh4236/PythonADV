## inplace:
- When inplace=True, the data is modified in place, which means it will return nothing and the dataframe is now updated.
- When inplace=False, which is the default, then the operation is performed and it returns a copy of the object.

```python
New.sort_values('Age in Yrs',inplace=True)
New
```

```python
New.sort_values('Age in Yrs',inplace=True,ascending=False,ignore_index=True)
New
```

## Dropna():
- The dropna() method removes the rows that contains NULL values.
- The dropna() method returns a new DataFrame object unless the inplace parameter is set to True, in that case the dropna() method does the removing in the original DataFrame instead.

```python
df=pd.read_csv('Emp_Records.csv')
df
```

```python
df.shape
```

```python
Drop=df.dropna()
Drop
```

```python
Drop.shape
```

## thresh: Specifies the number of NULL values required to remove the row or column.

```python
Drop=df.dropna(thresh=5)  # Min 5 non-null values in rows
Drop
```

```python
Drop=df.dropna(thresh=6)  # Min 6 non-null values in rows
Drop
```
