# Matplotlib:

- Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

- Matplotlib is open source and we can use it freely.

## How to import Matplotlib:
import matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,1,2,3])   # ([1,10])
y=np.array([2,4,6,8])   # ([0,20])
plt.plot(x,y)
plt.show()
```

## Plotting Without Line

```python
x_axis=np.array([0,5])
y_axis=np.array([3,7])
plt.plot(x_axis, y_axis, 'o')   # o is the parameter
plt.show()
```

## Multiple Points

```python
x_axis=np.array([1, 2, 6, 8])
y_axis=np.array([3, 8, 1, 10])
plt.plot(x_axis, y_axis)
plt.show()
```

## Default X-axis

```python
y_axis=np.array([3, 8, 1, 10, 5, 7])
plt.plot(y_axis)
plt.show()
```

## Markers
We can use the keyword argument marker to each point with a specified marker

```python
y_axis=np.array([3,6,9,12])
plt.plot(y_axis,marker='o')   # '*','.',
plt.show()
```

## Linestyle

```python
y_axis=np.array([3, 8, 1, 10])
plt.plot(y_axis, linestyle='dotted')   # linestyle='dashed'
plt.show()
```

### Shorter Syntax

- linestyle can be written as ls.

- dotted can be written as :'.'

- dashed can be written as '--'

```python
y_axis=np.array([3, 8, 1, 10])
plt.plot(y_axis, color='y')     # color='b','r'
plt.show()
```

## Line Width

```python
y_axis=np.array([3, 8, 1, 10])
plt.plot(y_axis, linewidth='11.10')
plt.show()
```

## Multiple Lines

```python
a1=np.array([3, 8, 1, 10])
a2=np.array([6, 2, 7, 11])
plt.plot(a1)
plt.plot(a2)
plt.show()
```

## Create Labels for a Plot
- With Pyplot, we can use the xlabel() and ylabel() functions to set a label for the x-axis and y-axis.

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y)
plt.xlabel("Pulse")
plt.ylabel("Rate")
plt.show()
```

## Create a Title for a Plot

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y)
plt.title("HeartRate")
plt.xlabel("Pulse")
plt.ylabel("Rate")
plt.show()
```

## Position the Title
- We can use the loc parameter in title() to position the title.

- Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y)
plt.title("HeartRate",loc='left')
plt.xlabel("Pulse")
plt.ylabel("Rate")
plt.show()
```

## Adding Grid Lines to a Plot

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y)
plt.title("HeartRate")
plt.xlabel("Pulse")
plt.ylabel("Rate")
plt.grid()      # plt.grid(axis='x')
plt.show()
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y)
plt.title("HeartRate")
plt.xlabel("Pulse")
plt.ylabel("Rate")
plt.grid(axis='y')
plt.show()
```

## Matplotlib Subplot
- Display Multiple Plots

```python
# plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,2,1)
plt.plot(x,y)

# plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

## Matplotlib Scatter
- Creating Scatter Plots

- The scatter() function plots one dot for each observation. 
- It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y) 
plt.show()
```

## Colors

```python
# plot-1
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='pink')

# plot-2
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='b')
plt.show()
```

## Matplotlib Bars
- We can provide heighy, width, color)

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 1, 10])
plt.bar(x,y)      # for horizontal bar use- barh
plt.show()
```

## Matplotlib Pie Charts

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show() 
```

## Labels

```python
y=np.array([35,25,25,15])
mylabels=['W','X','Y','Z']
plt.pie(y,labels=mylabels)
plt.show() 
```

## Explode

```python
y=np.array([35,25,25,15])
mylabels=['A','B','C','D']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show() 
```

## Shadow

```python
y=np.array([35,25,25,15])
mylabels=['A','B','C','D']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show() 
```

## Colors

```python
y=np.array([35,25,25,15])
mylabels=['A','B','C','D']
mycolors=['black','pink','b','red']
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

## Legend

```python
y=np.array([35,25,25,15])
mylabels=['A','B','C','D']
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

## Legend With Header

```python
y=np.array([35,25,25,15])
mylabels=['A','B','C','D']
plt.pie(y,labels=mylabels)
plt.legend(title="Alphabets")  # loc='left'
plt.show()
```

## Matplotlib Histograms

### Histogram
- A histogram is a graph showing frequency distributions.

- It is a graph showing the number of observations within each given interval.

```python
x=np.random.normal(100,20,100)
plt.hist(x)
plt.show() 
```
